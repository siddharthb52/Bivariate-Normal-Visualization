```python
import plotly.graph_objects as go
import numpy as np

# Prompt the user for mean, standard deviation, and covariance of both and Y

mean_x = float(input("Enter the mean of variable X: "))
std_x = float(input("Enter the standard deviation of variable X: "))
mean_y = float(input("Enter the mean of variable Y: "))
std_y = float(input("Enter the standard deviation of variable Y: "))
covariance = float(input("Enter the covariance between X and Y (Must be between -1 and 1): "))

# Compute the bounds of the graph using user input -- the center of the graph should have the x,y coordinates set to (mean_x, mean_y) as its center
# The minimum and maximum edges of the graph are calculated using the standard deviation values provided by the user. 
# The value of 3 * std was chosen with the empirical rule in mind. Accomodating 3 standard deviations should encapsulate 99.7% of the probability for a given normal distribution.

x_min = mean_x - 3 * std_x
x_max = mean_x + 3 * std_x
y_min = mean_y - 3 * std_y
y_max = mean_y + 3 * std_y

# Generate a grid of points within the bounds

x = np.linspace(x_min, x_max, 100)
y = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x, y)

# Compute the probability density function (PDF) of the bivariate normal distribution
Z = np.exp(-(X - mean_x) ** 2 / (2 * std_x ** 2) - (Y - mean_y) ** 2 / (2 * std_y ** 2) +
           covariance * (X - mean_x) * (Y - mean_y) / (std_x * std_y)) / (
            2 * np.pi * std_x * std_y * np.sqrt(1 - covariance ** 2))

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='magma')])

# Set axis labels
fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Probability Density'))

# Show the plot
fig.show()
```
